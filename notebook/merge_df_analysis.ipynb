```python
import re

# ì´ëª¨í‹°ì½˜ ì œê±° í•¨ìˆ˜ ì •ì˜
def remove_emojis(text):
    emoji_pattern = re.compile(
        "["
        u"\U0001F600-\U0001F64F"  # ê°ì • ì´ëª¨ì§€
        u"\U0001F300-\U0001F5FF"  # ê¸°í˜¸ ë° ë¬¼ì²´ ì´ëª¨ì§€
        u"\U0001F680-\U0001F6FF"  # êµí†µ ë° ê¸°ê³„ ì´ëª¨ì§€
        u"\U0001F700-\U0001F77F"  # ê¸°íƒ€ ì´ëª¨ì§€
        u"\U0001F780-\U0001F7FF"  # ë” ë§ì€ ê¸°íƒ€ ì´ëª¨ì§€
        u"\U0001F800-\U0001F8FF"  # ì¶”ê°€ ë²”ìœ„
        u"\U0001F900-\U0001F9FF"  # ì œìŠ¤ì²˜, ìŠ¤í¬ì¸  ì´ëª¨ì§€
        u"\U0001FA00-\U0001FA6F"  # ì˜¤ë¸Œì íŠ¸ ë° ì‹¬ë³¼
        u"\U0001FA70-\U0001FAFF"  # ì¶”ê°€ ì‹¬ë³¼
        u"\u2600-\u26FF"          # ê¸°íƒ€ ê¸°í˜¸
        u"\u2700-\u27BF"          # ë¬¸ì ê¸°í˜¸
        "]+", 
        flags=re.UNICODE
    )
    return emoji_pattern.sub(r'', text)

# 'ë¦¬ë·° í…ìŠ¤íŠ¸' ì»¬ëŸ¼ì—ì„œ ì´ëª¨í‹°ì½˜ ì œê±°
merged_df['ë¦¬ë·° í…ìŠ¤íŠ¸'] = merged_df['ë¦¬ë·° í…ìŠ¤íŠ¸'].astype(str).apply(remove_emojis)

# ê²°ê³¼ í™•ì¸
print(merged_df['ë¦¬ë·° í…ìŠ¤íŠ¸'].head(20))
```

```python
# ì´ëª¨í‹°ì½˜ì´ ë‚¨ì•„ìˆëŠ” í–‰ ì°¾ê¸°
def contains_emojis(text):
    emoji_pattern = re.compile(
        "["
        u"\U0001F600-\U0001F64F"  # ê°ì • ì´ëª¨ì§€
        u"\U0001F300-\U0001F5FF"  # ê¸°í˜¸ ë° ë¬¼ì²´ ì´ëª¨ì§€
        u"\U0001F680-\U0001F6FF"  # êµí†µ ë° ê¸°ê³„ ì´ëª¨ì§€
        u"\U0001F700-\U0001F77F"  # ê¸°íƒ€ ì´ëª¨ì§€
        u"\U0001F780-\U0001F7FF"  # ë” ë§ì€ ê¸°íƒ€ ì´ëª¨ì§€
        u"\U0001F800-\U0001F8FF"  # ì¶”ê°€ ë²”ìœ„
        u"\U0001F900-\U0001F9FF"  # ì œìŠ¤ì²˜, ìŠ¤í¬ì¸  ì´ëª¨ì§€
        u"\U0001FA00-\U0001FA6F"  # ì˜¤ë¸Œì íŠ¸ ë° ì‹¬ë³¼
        u"\U0001FA70-\U0001FAFF"  # ì¶”ê°€ ì‹¬ë³¼
        u"\u2600-\u26FF"          # ê¸°íƒ€ ê¸°í˜¸
        u"\u2700-\u27BF"          # ë¬¸ì ê¸°í˜¸
        "]+", 
        flags=re.UNICODE
    )
    return bool(emoji_pattern.search(text))

# ì´ëª¨í‹°ì½˜ì´ ë‚¨ì•„ìˆëŠ” í–‰ í•„í„°ë§
emoji_rows = merged_df[merged_df['ë¦¬ë·° í…ìŠ¤íŠ¸'].astype(str).apply(contains_emojis)]

# ê²°ê³¼ ì¶œë ¥
print(emoji_rows)
```

```python
# í•„ìš”í•œ ë¼ì´ë¸ŒëŸ¬ë¦¬ ë¶ˆëŸ¬ì˜¤ê¸°
from konlpy.tag import Okt
from collections import Counter
import requests

# ë¶ˆìš©ì–´ ë¦¬ìŠ¤íŠ¸ ë‹¤ìš´ë¡œë“œ
url = "https://raw.githubusercontent.com/stopwords-iso/stopwords-ko/master/stopwords-ko.txt"
response = requests.get(url)
stopwords = response.text.split("\n")  # ë¶ˆìš©ì–´ ë¦¬ìŠ¤íŠ¸ ë¡œë“œ

# í˜•íƒœì†Œ ë¶„ì„ê¸° ì„ ì–¸
okt = Okt()

# ëŒ€í•™êµë³„ ë‹¨ì–´ ë¹ˆë„ ë¶„ì„ í•¨ìˆ˜ (Top 50 ì¶”ì¶œ)
def get_top_50_words_by_university(df, text_column, university_column):
    result = {}
    
    # ëŒ€í•™ë³„ë¡œ ë°ì´í„° ê·¸ë£¹í™”
    for university, group in df.groupby(university_column):
        # í•´ë‹¹ ëŒ€í•™ì˜ ëª¨ë“  ë¦¬ë·° í…ìŠ¤íŠ¸ í•©ì¹˜ê¸°
        text = " ".join(group[text_column].dropna())
        
        # í˜•íƒœì†Œ ë¶„ì„ í›„ ëª…ì‚¬ë§Œ ì¶”ì¶œ
        tokens = okt.nouns(text)
        
        # ë¶ˆìš©ì–´ ì œê±°
        tokens = [word for word in tokens if word not in stopwords and len(word) > 1]
        
        # ë‹¨ì–´ ë¹ˆë„ ê³„ì‚°
        word_freq = Counter(tokens)
        
        # ìƒìœ„ 50ê°œ ë‹¨ì–´ ì¶”ì¶œ
        result[university] = word_freq.most_common(50)
    
    return result

# í•¨ìˆ˜ ì‹¤í–‰
university_top_50_words = get_top_50_words_by_university(merged_df, 'ë¦¬ë·° í…ìŠ¤íŠ¸', 'ëŒ€í•™êµ ì´ë¦„')

# ê²°ê³¼ ì¶œë ¥
for university, words in university_top_50_words.items():
    print(f"ëŒ€í•™êµ: {university}")
    print("ìƒìœ„ 50 ë‹¨ì–´:", words)
    print()
```
